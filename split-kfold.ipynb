```python
from sklearn.model_selection import (TimeSeriesSplit, KFold, ShuffleSplit,
                                     StratifiedKFold, GroupShuffleSplit,
                                     GroupKFold, StratifiedShuffleSplit)
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
np.random.seed(1338)
cmap_data = plt.cm.Set2
cmap_cv = plt.cm.Set2
n_splits = 5
COLOR_BINARY = [
    [x/255 for x in [217,95,2]],
    [x/255 for x in [117,112,179]],
]
COLOR_SCHEME = [ 
        [x/255 for x in [166,206,227]],  
        [x/255 for x in [31,120,180]],
        [x/255 for x in [51,160,44]],
      ]
from matplotlib.colors import ListedColormap
cm = ListedColormap(np.array(COLOR_SCHEME))
cm_bin = ListedColormap(np.array(COLOR_BINARY))
```

```python
# Generate the class/group data
n_points = 100
X = np.random.randn(100, 10)

percentiles_classes = [.2, .3, .5]
y = np.hstack([[ii] * int(100 * perc)
               for ii, perc in enumerate(percentiles_classes)])

# Evenly spaced groups repeated once
groups = np.hstack([[ii] * 10 for ii in range(10)])
```

```python
def plot_cv_indices(cv, X, y, group, ax, n_splits, lw=15):
    """Create a sample plot for indices of a cross-validation object."""

    # Generate the training/testing visualizations for each CV split
    for ii, (tr, tt) in enumerate(cv.split(X=X, y=y, groups=group)):
        # Fill in indices with the training/test groups
        indices = np.array([np.nan] * len(X))
        indices[tt] = 1
        indices[tr] = 0

        # Visualize the results
        ax.scatter(range(len(indices)), [ii + .5] * len(indices),
                   c=indices, marker='_', lw=lw, cmap=cm_bin,
                   vmin=-.2, vmax=1.2)

    # Plot the data classes and groups at the end
    ax.scatter(range(len(X)), [ii + 1.5] * len(X),
               c=y, marker='_', lw=lw, cmap=cm)

    ax.scatter(range(len(X)), [ii + 2.5] * len(X),
               c=group, marker='_', lw=lw, cmap=cm)

    # Formatting
    yticklabels = list(range(1,n_splits+1)) + ['class', 'group']
    ax.set(yticks=np.arange(n_splits+2) + .5, yticklabels=yticklabels,
            ylabel="CV iteration",
           ylim=[n_splits+1.2, -0.2], xlim=[0, 100])
    ax.set_xlabel('Sample index', fontsize=16)
    ax.set_ylabel('CV iteration', fontsize=16)
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.set_title('Stratified 5-fold with randomly shuffled data', fontsize=16)
    return ax
```

```python
FIG_DIR = 'figures/'
fig, ax = plt.subplots(figsize=(8,4))
cv = StratifiedKFold(n_splits, shuffle=True)
plot_cv_indices(cv, X, y, groups, ax, n_splits)

ax.legend([Patch(color=COLOR_BINARY[0]), Patch(color=COLOR_BINARY[1])],
          ['Testing set', 'Training set'], loc=(1.02, 0.5), fontsize=16)
# Make the legend fit
plt.tight_layout()
fig.subplots_adjust(right=.7)
filename = FIG_DIR + 'stratified-5fold.png'
plt.savefig(filename, dpi=300)
!convert -trim $filename $filename
```
